```python
corpus=["I like playing Call of Duty Mobile","I always like teaming up with my friends","I prefer Battle royale mode"]
nlp= spacy.load("en_core_web_sm")

def pre_processor(text):
    doc=nlp(text)

    filtered_tokens=[]
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    return filtered_tokens

pre_processor("I hate playing with fucking randoms")
```

```python
#Use list comprehension to iterate over the corpus

pre_processed_corpus=[' '.join(pre_processor(sentence)) for sentence in corpus ]
pre_processed_corpus
```

```python
v= CountVectorizer(ngram_range=(1,2))
v.fit(pre_processed_corpus)
v.vocabulary_
```

```python
import pandas as pd
import json

data = []
with open("/content/drive/MyDrive/News_Category_Dataset_v3.json", 'r') as f:
    for line in f:
        try:
            data.append(json.loads(line))  # Parse each line as a JSON object
        except json.JSONDecodeError:
            pass  # Ignore lines that aren't valid JSON (like the trailing data)

df = pd.DataFrame(data)
print(df.shape)
df.head()
```

```python
df.category.value_counts()
```

```python
# using undersampling technique since data categories are underbalanced
minority_class_size = df['category'].value_counts().min()
# Undersample each class
df_balanced = df.groupby('category').apply(lambda x: x.sample(minority_class_size)).reset_index(drop=True)
# 3. Check the new class distribution
print(df_balanced['category'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(df_balanced['headline'],df_balanced['category'],test_size=0.2,random_state=2022,stratify=df_balanced['category'])
```

```python
y_train.value_counts()
```

```python
x_test.value_counts()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report

pipe=Pipeline([('vectorizer_bow',CountVectorizer()),('Multi NB',MultinomialNB())])

pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print(classification_report(y_test,y_pred))
```
